## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
def convert_register(name, force = None):
    if force is None:
        if "_" in name:
            source = "snake"
        else:
            source = "pascal"
    else:
        source = force

    if source == "snake":
        parts = name.split("_")
        return "".join(word.capitalize() for word in parts)
    
    elif source == "pascal":
        result = ""
        for ch in name:
            if ch.isupper():
                if result:
                    result += "_"
                result += ch.lower()
            else:
                result += ch
        return result
    else:
        raise ValueError("force must be 'None', 'snake' or 'pascal'")
    
print(convert_register("otus_course"))
print(convert_register("PythonIsTheBest"))

```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def valid_date(date):
    parts = date.split(".")
    
    if len(parts) != 3:
        return False
    
    day_str, month_str, year_str = parts

    if not(day_str.isdigit() and month_str.isdigit() and year_str.isdigit()):
        return False
    
    day = int(day_str)
    month = int(month_str)
    year = int(year_str)

    if(month < 1 or month > 12):
        return False


    days_in_month = [
        31,28,31,30,31,30,31,31,30,31,30,31
    ]

    #Проверка на високосный год
    is_leap = (year % 400 == 0) or (year % 4 ==0 and year %100 !=0)

    if is_leap:
        days_in_month[1] = 29

    return 1 <= day <= days_in_month[month-1]

print(valid_date("29.02.2000"))
print(valid_date("29.02.2001"))
print(valid_date("31.04.1962"))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_simple_number(number):
    if number < 2:
        return True
    
    for i in range(2, int(number ** 0.5)+ 1):
        if number % i ==0:
            return False
        
    return True
    
print(is_simple_number(17))
print(is_simple_number(20))
print(is_simple_number(23))
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def normalize_name(value: str) -> str:
    value = value.strip()
    if value == "":
        return ""
    #проверка первой буквы
    first = value[0].upper()
    rest = value[1:]
    return first+rest

def is_letters_only(value: str) -> bool:
    for ch in value:
        if not ch.isalpha():
            return False
    return True

def read_age() -> int:
    while True:
        raw = input("Введите возраст от 18 до 60: ").strip()
        if(raw == ""):
            #окончание ввода
            return -1
        if not raw.isdigit():
            print("Возраст должен быть числом")
            continue
        age = int(raw)
        if 18 <= age <=60:
            return age
        
        print("Возраст не в диапазное 18-60 лет")

def read_id(existing_ids: set[str]) -> str:
    while True:
        raw = input("ID пользователя в виде челого числа : ").strip()
        if raw == "":
            return ""
        if not raw.isdigit():
            print("ID должен быть числом")
            continue

        user_id = raw.zfill(8)
        if user_id in existing_ids:
            print("Такой ID уже есть в таблице, выберите другой")
            continue

        return user_id      
    
def print_user_table(users: dict[str, tuple[str, str, int]]) : 
    if not users:
        print("Пользователей в таблице нет")
        return
    
    headers = ("ID" , "Имя", "Фамилия", "Возраст")

    id_width = max(len(headers[0]) , max(len(uid) for uid in users))
    name_width = max(len(headers[1]) , max(len(data[0]) for data in users.values()))
    surname_width = max(len(headers[2]) , max(len(data[1]) for data in users.values()))
    age_width = len(headers[3])
    #заголовок
    print(f"{headers[0]:<{id_width}} {headers[1]:<{name_width}} {headers[2]:<{surname_width}} {headers[3]:<{age_width}}")
    #форматирование по ширине
    print(f"{'-'*id_width} {'-'*name_width} {'-'*surname_width} {'-'*age_width}")

    for uid in sorted(users.keys()):
        name, surname, age = users[uid]
        print(f"{uid:<{id_width}} {name:<{name_width}} {surname:<{surname_width}} {age:<{age_width}}")


def main():
    users : dict[str, tuple[str,str,int]] = {}
    used_ids : set[str] = set()

    while True:
        name_input = input("Имя: ").strip()
        if name_input == "":
            break
        
        surname_input = input("Фамилия: ").strip()
        if surname_input == "":
            break

        name = normalize_name(name_input)
        surname = normalize_name(surname_input)

        if not is_letters_only(name) or not is_letters_only(surname):
            print("Имя и фамилия должны содержать только буквы")
            continue
        age = read_age()
        #сигнал хавершения работы
        if(age == -1):
            break

        user_id = read_id(used_ids)
        if user_id == "":
            break
        users[user_id] = (name, surname,age)
        used_ids.add(user_id)
        print("Пользователь успешно добавлен")

    print("Словарь пользователей")
    print_user_table(users)


if __name__ == "__main__":
    main()






  
    

```
